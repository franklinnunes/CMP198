```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import LeaveOneOut
import shap
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, log_loss, confusion_matrix, make_scorer
from sklearn.model_selection import StratifiedKFold
import lime
import lime.lime_tabular
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
import lightgbm as lgb
```

# Data read, convert condition to binary and feature normalization

```python
data = pd.read_csv("Lung_Hsapiens_GSE81089.csv")
X = data.drop(columns=['Sample', 'Condition'])
y = data['Condition'].map({'Tumor': 1, 'Normal': 0})
scaler = MinMaxScaler()
X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
```

```python
X.shape[1]/2
```

# Filter feature selection SelectKBest with f_classif score function

```python
from sklearn.feature_selection import SelectKBest, f_classif

selector = SelectKBest(score_func=f_classif, k=19987)
X_new = selector.fit_transform(X, y)

selected_indices = selector.get_support(indices=True)
selected_features = X.columns[selected_indices]

X_new = pd.DataFrame(X_new, columns=selected_features)

print(X_new.shape)
```

# Train test split 80% train/ 20% test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.2, stratify=y)
```

# Bayesian optimization - RF

```python
from hyperopt import fmin, tpe, hp, STATUS_OK

space_rf = {
    'n_estimators': hp.quniform('n_estimators', 50, 500, 10),
    'max_depth': hp.quniform('max_depth', 2, 20, 1),
    'min_samples_split': hp.quniform('min_samples_split', 2, 20, 1),
    'min_samples_leaf': hp.quniform('min_samples_leaf', 1, 20, 1),
    'bootstrap': hp.choice('bootstrap', [True, False])
}

def objective_rf(params):
    params['n_estimators'] = int(params['n_estimators'])
    params['max_depth'] = int(params['max_depth'])
    params['min_samples_split'] = int(params['min_samples_split'])
    params['min_samples_leaf'] = int(params['min_samples_leaf'])
    
    rf_model = RandomForestClassifier(**params)
    rf_model.fit(X_train, y_train)
    y_pred = rf_model.predict(X_test)
    score = accuracy_score(y_test, y_pred)
    return {'loss': -score, 'status': STATUS_OK}

best_params_rf = fmin(objective_rf, space_rf, algo=tpe.suggest, max_evals=100)
print(best_params_rf)
```

# Bayesian optimization - DT

```python
space_dt = {
    'max_depth': hp.quniform('max_depth', 2, 20, 1),
    'min_samples_split': hp.quniform('min_samples_split', 2, 20, 1),
    'min_samples_leaf': hp.quniform('min_samples_leaf', 1, 20, 1),
}

def objective_dt(params):
    params['max_depth'] = int(params['max_depth'])
    params['min_samples_split'] = int(params['min_samples_split'])
    params['min_samples_leaf'] = int(params['min_samples_leaf'])
    
    dt_model = DecisionTreeClassifier(**params)
    dt_model.fit(X_train, y_train)
    y_pred = dt_model.predict(X_test)
    score = accuracy_score(y_test, y_pred)
    return {'loss': -score, 'status': STATUS_OK}

best_params_dt = fmin(objective_dt, space_dt, algo=tpe.suggest, max_evals=100)
print(best_params_dt)
```

# Bayesian optimization - LightGBoost

```python
space = {
    'max_depth': hp.quniform('max_depth', 2, 20, 1), #max_depth precisa ser int
    'learning_rate': hp.loguniform('learning_rate', -5, -2),
    'subsample': hp.uniform('subsample', 0.5, 1)
}


def objective(params):
    params['max_depth'] = int(params['max_depth']) 
    lgb_model = lgb.LGBMClassifier(**params, verbosity=-1)
    lgb_model.fit(X_train, y_train)
    y_pred = lgb_model.predict(X_test)
    score = accuracy_score(y_test, y_pred)
    return {'loss': -score, 'status': STATUS_OK}

best_params = fmin(objective, space, algo=tpe.suggest, max_evals=100)
print("Best set of hyperparameters: ", best_params)
```

```python
rfc  = RandomForestClassifier(max_depth= 9, min_samples_leaf= 12, min_samples_split= 15, n_estimators= 80)
lgbc  = lgb.LGBMClassifier(learning_rate= 0.051142965005344494, max_depth= 5, subsample= 0.6497325384465419, verbosity=-1)
dtc  = DecisionTreeClassifier(max_depth= 19, min_samples_leaf= 13, min_samples_split= 16)
```

# Random Forest loocv metrics

```python
accuracy_list = []
precision_list = []
recall_list = []
f1_list = []
roc_auc_list = []
logloss_list = []
sensitivity_list = []
specificity_list = []
lr_plus_list = []
lr_minus_list = []

for i in range(11):
    loo = LeaveOneOut()
    y_true_accum = []
    y_pred_accum = []
    y_pred_proba_accum = []

    for train_index, test_index in loo.split(X_new):
        X_train_fold, X_test_fold = X_new.iloc[train_index], X_new.iloc[test_index]
        y_train_fold, y_test_fold = y.iloc[train_index], y.iloc[test_index]

        rfc.fit(X_train_fold, y_train_fold)
        y_pred_fold = rfc.predict(X_test_fold)
        y_pred_proba_fold = rfc.predict_proba(X_test_fold)[:, 1]

        y_true_accum.extend(y_test_fold)
        y_pred_accum.extend(y_pred_fold)
        y_pred_proba_accum.extend(y_pred_proba_fold)

    #arrays numpy
    y_true_accum = np.array(y_true_accum)
    y_pred_accum = np.array(y_pred_accum)
    y_pred_proba_accum = np.array(y_pred_proba_accum)

    accuracy = accuracy_score(y_true_accum, y_pred_accum)
    precision = precision_score(y_true_accum, y_pred_accum)
    recall = recall_score(y_true_accum, y_pred_accum)
    f1 = f1_score(y_true_accum, y_pred_accum)
    roc_auc = roc_auc_score(y_true_accum, y_pred_proba_accum)
    logloss = log_loss(y_true_accum, y_pred_proba_accum)

    conf_mat = confusion_matrix(y_true_accum, y_pred_accum)
    tn, fp, fn, tp = conf_mat.ravel()
    sensitivity = tp / (tp + fn)
    specificity = tn / (tn + fp)
    lr_plus = sensitivity / (1 - specificity)
    lr_minus = (1 - sensitivity) / specificity

    accuracy_list.append(accuracy)
    precision_list.append(precision)
    recall_list.append(recall)
    f1_list.append(f1)
    roc_auc_list.append(roc_auc)
    logloss_list.append(logloss)
    sensitivity_list.append(sensitivity)
    specificity_list.append(specificity)
    lr_plus_list.append(lr_plus)
    lr_minus_list.append(lr_minus)

accuracy_mean = np.mean(accuracy_list)
precision_mean = np.mean(precision_list)
recall_mean = np.mean(recall_list)
f1_mean = np.mean(f1_list)
roc_auc_mean = np.mean(roc_auc_list)
logloss_mean = np.mean(logloss_list)
sensitivity_mean = np.mean(sensitivity_list)
specificity_mean = np.mean(specificity_list)
lr_plus_mean = np.mean(lr_plus_list)
lr_minus_mean = np.mean(lr_minus_list)

print(f'Mean Accuracy: {accuracy_mean:.6f}')
print(f'Mean Precision: {precision_mean:.6f}')
print(f'Mean Recall: {recall_mean:.6f}')
print(f'Mean F1-score: {f1_mean:.6f}')
print(f'Mean ROC AUC: {roc_auc_mean:.6f}')
print(f'Mean Log Loss: {logloss_mean:.6f}')
print(f'Mean Sensitivity: {sensitivity_mean:.6f}')
print(f'Mean Specificity: {specificity_mean:.6f}')
print(f'Mean LR+: {lr_plus_mean:.6f}')
print(f'Mean LR-: {lr_minus_mean:.6f}')
```

# Random forest feature_importances_

```python
def feature_importance(X, y,model, n_runs=11, test_size=0.2):
    importances = np.zeros(X.shape[1])

    for i in range(n_runs):
        model.fit(X_train, y_train)

        importances += model.feature_importances_

    importances /= n_runs

    feature_importance = pd.DataFrame({
        'Feature': X.columns,
        'Importance': importances
    })

    feature_importance = feature_importance.sort_values(by='Importance', ascending=False)

    return feature_importance
```

```python
feature_importance_df = feature_importance(X_new, y, rfc)
top_10_features = feature_importance_df.head(10)
```

# Explain tumor samples from test dataset with LIME (Local Interpretable Model-Agnostic Explanations)

```python
import numpy as np
import lime.lime_tabular
import matplotlib.pyplot as plt

explainer = lime.lime_tabular.LimeTabularExplainer(
    training_data=np.array(X_test),
    feature_names=X_test.columns,
    class_names=['Normal', 'Tumor'],
    mode='classification'
)

#apenas incluir as amostras tumorais
tumor_samples = X_test[rfc.predict(X_test) == 1]

lime_features_list = []

for i in range(tumor_samples.shape[0]):
    sample = tumor_samples.iloc[i]
    
    explanation = explainer.explain_instance(
        data_row=np.array(sample),
        predict_fn=rfc.predict_proba
    )
    
    plt.figure(figsize=(8, 6))
    explanation.as_pyplot_figure()
    plt.title(f"LIME Explanation for Tumor Sample {i+1}")
    plt.show()

    #arquivo de txt
    lime_features = [x[0] for x in explanation.as_list()]
    lime_features_list.append(f"Sample {i+1}: " + ", ".join(lime_features) + "\n")


with open("lime_features.txt", "w") as f:
    f.writelines(lime_features_list)
```
